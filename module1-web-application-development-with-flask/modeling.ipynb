```python
import pandas as pd
from nba_api.stats.endpoints import playercareerstats, franchisehistory, playoffpicture,leagueleaders
from nba_api.stats.static import players
```

Lets look at what player you would be in the nba based on the following questions:
1. Three points taken vs total shots taken... Do you go for big wins or little wins?
2. Mins played... Do you stay for the whole party or irish exit often?
3. Assists per game...Would you rather have your team win or you have all the glory!
4. Offensive and defensive rebounds...Do you join the mosh pit or take a seat?

```python
seasons = ['2015-16','2016-17','2017-18','2018-19']
```

```python
df = pd.DataFrame()
for season in seasons:
    df = df.append(leagueleaders.LeagueLeaders(season=season).get_data_frames()[0])
```

```python
df = df.loc[df['RANK'] <= 30]
```

```python
played_ids = df['PLAYER_ID'].unique()
```

```python
features = ['RANK','GP','MIN','FG_PCT','FG3_PCT','REB','STL','BLK']
```

```python
target = ['PLAYER']
```

```python
X = df[features]
y = df[target]
```
